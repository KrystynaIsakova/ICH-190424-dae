Совмещение двух графиков

Изобразите на одной системе координат графики функций $y=sin(x)$ и $y=log(x+1)$. Используйте две оси Y: одну для первого графика, и вторую для другого. 


````python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
import pandas as pd
````

````python
x = np.linspace(1e-4, 50, 100)
y1 = np.sin(x)
y2 = np.log1p(x)

plt.figure(figsize = (9,9))
plt.grid()
ax1 = plt.subplot()
ax1.set_xlabel('x')
ax1.set_ylabel('$\sin(x)$', color='red')
ax1.plot(x,y1, color='red')
ax1.tick_params(axis='y', labelcolor='red')

ax2 = ax1.twinx()
ax2.plot(x,y2, color = 'black')
ax2.set_ylabel('$\log(x)$', color='black')
ax2.tick_params(axis='y', labelcolor='black')

plt.show()
````
