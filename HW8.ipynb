Отобразите первые 5 строчек датафрейма. Визуализируйте эффект типа события. Для этого постройте barplot, отображающий среднее значение display для каждого события в момент времени timepoint=5. Не забудьте назвать график.


```python
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
import matplotlib.pyplot as plt
```

```python
# решение
import seaborn as sns
fmri = sns.load_dataset('fmri')
fmri.head(5)

sns.barplot(data=fmri.query("timepoint == 5"), x='event', y='signal')
plt.show()
```

Изучите влияние области на результаты томографии. Отобразите с помощью pointplot средний сигнал в каждый момент времени в зависимости от области снятия показателей. Не забудьте назвать график.


```python
sns.pointplot(data=fmri, x='timepoint', y='signal', hue='region')
plt.show()
```

Интерактивная визуализация с помощью FacetGrid. Используя FacetGrid, постройте отдельные lineplot для каждого субъекта, каждый график должен показывать связь между временем и сигналом. Задача со звёздочкой: подписать каждый график.

```python
g = sns.FacetGrid(fmri, col='subject', col_wrap=5, height=2, sharey=False)
g.map(sns.lineplot, 'timepoint', 'signal')
g.set_titles('Субъект {col_name}')
plt.show()
```
