Задание 1: Расширенная диаграмма рассеяния
Создать диаграмму рассеивания с дополнительной сложностью для визуализации взаимосвязи между двумя переменными. Используйте набор данных tips, доступный в Plotly Express. Постройте график total_bill на оси x и tip на оси y. Раскрасьте точки по времени (обед или ужин). Добавьте на график линию тренда и настройте размер маркеров в зависимости от размера вечеринки.


```python
import plotly.express as px
```

```python
tips = px.data.tips()
```

```python
fig = px.scatter(tips,
          x = 'total_bill',
          y = 'tip',
        color = 'time',
        trendline='ols',
        size = 'size')
fig.show()
```

Задача 2: Создать гистограмму с расширенными возможностями настройки. Используйте набор данных tips. Создайте гистограмму для total_bill. Установите число бинов на 20 и цвет по времени.


```python
fig = px.histogram(tips,
             x = 'total_bill',
            nbins=20,
            color = 'time')
fig.show()
```

Задача 3: Создайте линейный график с несколькими линиями для сравнения тенденций. Используйте набор данных gapminder, доступный в Plotly Express. Отметьте год на оси x и gdpPercap на оси y. Отфильтруйте данные, чтобы включить в них только страны "Канада" и "Германия". Используйте разные стили линий для каждой страны.


```python
gapminder = px.data.gapminder().query('country in ["Canada", "Germany"]')

fig = px.line(gapminder,
             x = 'year',
             y = 'gdpPercap',
             color ='country')
fig.show()
```

Задача 4: Создать тепловую карту для визуализации корреляции между несколькими переменными с помощью интерактивных функций. Используйте набор данных iris, доступный в Plotly Express. Рассчитайте корреляционную матрицу и визуализируйте ее с помощью тепловой карты. Добавьте наведенную информацию, показывающую значения корреляции.


```python
iris = px.data.iris()
```

```python
import numpy as np
corr = np.round(iris.corr(numeric_only=True),2)
```

```python
fig = px.imshow(corr,
               text_auto = True)
fig.show()
```
