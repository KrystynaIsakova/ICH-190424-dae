 Простые операции
 ---
Рассмотрим простые операции над DataFrame. Очень сильно они будут похожи на аналогичные операции для Series.


```python
df1 = pd.DataFrame(np.random.uniform(size=(10, 4)),
                   columns=['A', 'B', 'C', 'D'])
display(df1)  # создадим таблицу из массива случайных чисел.

df2 = pd.DataFrame(np.random.uniform(size=(7, 3)),
                   columns=['A', 'B', 'C'])
display(df2)  # и еще одну.

display(df1 + df2)

display(2 * df1 + 3)  # да, так можно!

display(np.sin(df1))  # и даже так

display(np.exp(df1['A']))  # можем применить для одного столбца (помним, что это Series)

display(df1.cumsum())  # старые знакомые – кумулятивные суммы 
```

Функции pd.merge и df.join
--
Пример 1. В обеих таблицах ключи повторяются.


```python
left = pd.DataFrame({'key': ['A', 'A'],
                     'lval': [1, 2]})
right = pd.DataFrame({'key': ['A', 'A'],
                      'rval': [4, 5]})
display(left)
display(right)
```

```python
pd.merge(left, right, on='key')
```

Пример 2. В таблицах ключи не повторяются.


```python
left = pd.DataFrame({'key': ['A', 'B'],
                     'lval': [1, 2]})
right = pd.DataFrame({'key': ['A', 'B'],
                      'rval': [4, 5]})
```

```python
pd.merge(left, right, on='key')
```

Пример 3. Посмотрим на различные типы объединения. Сооздадим и напечатаем две таблицы.


```python
left = pd.DataFrame({'lkey': ['A', 'B', 'C', 'A'],
                     'value': range(4)})
right = pd.DataFrame({'rkey': ['A', 'B', 'D', 'B'],
                      'value': range(4, 8)})
display(left)
display(right)
```

Внешнее слияние — используются ключи из объединения списков ключей. Иначе говоря, используются ключи, которые есть хотя бы в одной из таблиц. Если в другой таблице таких ключей нет, то ставятся пропуски.

```python
pd.merge(left, right,
         left_on='lkey', right_on='rkey', how='outer')
```

```python
pd.merge(left, right,
         left_on='lkey', right_on='rkey', how='outer').dtypes
```

```python
pd.merge(left, right,
         left_on='lkey', right_on='rkey', how='outer')['value_x'].values
```

Внутреннее слияние — используются ключи из пересечения списков ключей. Иначе говоря, используются ключи, которые присутствуют в обеих таблицах.


```python
pd.merge(left, right,
         left_on='lkey', right_on='rkey', how='inner')
```

Объединение по ключам левой таблицы. Не используются ключи, которые есть в правой таблицы, но которых нет в левой. Если в правой таблице каких-то ключей нет, то ставятся пропуски.


```python
pd.merge(left, right,
         left_on='lkey', right_on='rkey', how='left')
```

Объединение по ключам правой таблицы – аналогично.
Выполним внтуреннее объединение и установим ключ качестве индекса


```python
pd.merge(left, right, left_on='lkey', right_on='rkey', how='inner') \
        .set_index('lkey')[['value_x', 'value_y']]
```

Ту же операцию можно выполнить с помощью join


```python
left.set_index('lkey') \
    .join(right.set_index('rkey'), rsuffix='_r', how='inner')
```

Группировка
==
Группировка выполняется функцией


```python
df = pd.DataFrame({
    'Животное' : ['Котик', 'Песик', 'Котик', 'Песик',
                  'Котик', 'Песик', 'Котик', 'Песик'],
    'Цвет шерсти' : ['белый', 'белый', 'коричневый', 'черный',
                     'коричневый', 'коричневый', 'белый', 'черный'],
    'Рост' : np.random.gamma(12, scale=3,size=8),
    'Длина хвостика' : np.random.gamma(10,size=8)
})
df
```

Пример 1.
Если все котики встанут друг на друга, то какой их суммарный рост? А у песиков? А какова суммарная длина хвостиков у котиков и у песиков?
Группировка по одной колонке и последующее применение операции суммирования:


```python
df.groupby('Животное').sum()
```

Пример 2.
Теперь предположим, что котики и песики встают только на представителей своего вида и своего цвета шерсти. Что тогда будет?
Группировка по двум колонкам и последующее применение операции суммирования


```python
df.groupby(['Животное', 'Цвет шерсти']).sum()
```

```python
df.groupby(['Животное', 'Цвет шерсти']).sum().index
```

```python
np.where(df==5, 0, df)
```

```python
df.groupby('Животное')[['Длина хвостика','Рост']].agg({'Длина хвостика': 'sum',
                           'Рост' :'max'})
```
