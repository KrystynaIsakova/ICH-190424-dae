```python
# Линейная зависимость
x = np.linspace(0, 10, 50)
y1 = x
# Квадратичная зависимость
y2 = [i**2 for i in x]
# Построение графиков
plt.figure(figsize=(9, 9))
plt.subplot(2, 1, 1)
plt.plot(x, y1)               # построение графика
plt.title(r"Зависимость $y_1 = x$") # заголовок
plt.ylabel(r"$y_1$", fontsize=14) # ось ординат
plt.grid(True)                # включение отображение сетки
plt.subplot(2, 1, 2)
plt.plot(x, y2)               # построение графика
plt.title(r"Зависимость $y_2 = x^2$") # заголовок
plt.xlabel("x", fontsize=14)  # ось абсцисс
plt.ylabel(r"$y_2$", fontsize=14) # ось ординат
plt.grid(True)                # включение отображение сетки
plt.show()
```

Два подграфика одного рисунка могут также иметь общую ось Х, но разные оси Y. Для создания второго подграфика в этом случае нужно использовать метод twinx(). 

```python
month = ['янв','фев','март','апр','май','июнь','июль','авг','сен','окт','ноя','дек']
temp = [1.8, 1.8, 4.8, 9.0, 14.0, 18.8, 19.8, 19.6, 15.0, 10.0, 5.0, 2.0]
prec = [56, 41, 53, 42, 60, 67, 81, 62, 56, 49, 48, 54]

plt.figure()
ax1 = plt.subplot()

ax1.set_xlabel('Month')
ax1.set_ylabel(r'Temperature, $\degree$C', color='red')
ax1.plot(month, temp, color='red')
ax1.tick_params(axis='y', labelcolor='red')

ax2 = ax1.twinx()
ax2.set_ylabel('Precipitation, mm', color='blue')
ax2.plot(month, prec, color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

plt.show()
```

Еще одна возможность отобразить разную информацию на одном рисунке – использование вложенных графиков. Для этого нужно воспользоваться функцией axes(), которая создает оси в заданном прямоугольнике. При этом вся площадь рисунка описывается прямоугольником [0, 0, 1, 1]. 

plt.axes([x, y, width, height])

```python
# Основной график
plt.axes([0, 0, 1, 1])
plt.plot(range(10), color='red')
plt.title('Основной график')

# Вложенный график
plt.axes([0.2, 0.6, 0.25, 0.25])
plt.plot(range(10), color='blue')
plt.title('Вложенный график')
plt.show()
```

3D графики.
--

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y) #создаем матрицу из всех возможных комбинаций X и Y
Z = np.sin(np.sqrt(X**2 + Y**2))

surf = ax.plot_surface(X, Y, Z, cmap='viridis')
fig.colorbar(surf)
ax.can_zoom()
plt.show()
```

2. На одной системе координат постройте два трехмерных графика: Z=X*Y и Z=-X*Y.

```python
%matplotlib widget
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
Z = -X*Y
surf = ax.plot_surface(X, Y, Z, cmap='Oranges')
Z = X*Y
surf = ax.plot_surface(X, Y, Z, cmap='viridis')
ax.can_zoom()
plt.show()
```
